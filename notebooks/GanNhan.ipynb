```python
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("/home/sonn/Son/Workspace/IR_PhanLoaiNuocCam/IR/data/processed/wn.csv")
```

```python
tenmau = data["TenMau"]
```

```python
ma = []
for mau in tenmau:
    ma.append(mau.replace("/", "")[0:2])
```

```python
ma
```

```python
data2 = data.drop(["TenMau"], axis=1)
```

```python
data2["Ma"] = ma
```

```python
data2.to_csv("/home/sonn/Son/Workspace/IR_PhanLoaiNuocCam/IR/data/processed/data_encoded.csv")
```

```python
import os
import fnmatch
import pandas as pd
```

```python
def list_files(directory, extensions):
    matched_files = []
    for root, dirs, files in os.walk(directory):
        for ext in extensions:
            for filename in fnmatch.filter(files, f"*{ext}"):
                matched_files.append(os.path.join(root, filename))
    return matched_files

directory = '/home/sonn/Son/Workspace/IR_PhanLoaiNuocCam/IR/data/raw/IR.3.2024/IR.3.2024/TM-19.04.2024'
extensions = ['.csv', '.xlsx']
# extensions = ['.xlsx']
files = list_files(directory, extensions)
```

```python
tenfile = []

for file in files:
    tenfile.append(str(file).replace(directory, ""))
```

```python
tenfile
```

```python
tenmau = []

for i in range(len(tenfile)):
    i1 = str(tenfile[i]).replace(".csv", "")
    i2 = str(i1).replace(".xlsx", "")
    tenmau.append(i2)
```

```python
goc = pd.read_csv("/home/sonn/Son/Workspace/IR_PhanLoaiNuocCam/IR/data/processed/wn.csv")
```

```python
for k in range(len(tenfile)):
    if ".csv" in tenfile[k]:
        filecsv = pd.read_csv(files[k])
        goc[tenmau[k]] = filecsv["Intensity"]
    else:
        fileex = pd.read_excel(files[k])
        goc[tenmau[k]] = fileex["Intensity"]
        
```

```python
goc.to_csv("/home/sonn/Son/Workspace/IR_PhanLoaiNuocCam/IR/data/processed/wn.csv", index=False)
```

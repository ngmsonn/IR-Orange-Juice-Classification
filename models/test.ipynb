```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7])
x = x.reshape(-1, 1)
y = np.array([2,5,7,9,10,12,14])
y = y.reshape(-1, 1)
```

```python
from sklearn.neural_network import MLPRegressor

# Train your model
mlp = MLPRegressor(hidden_layer_sizes=(5,), max_iter=1000)
mlp.fit(x, y)
```

```python

# Coefficients and intercepts
for i, coef in enumerate(mlp.coefs_):
    print(f'Layer {i} weights:')
    print(coef)
    
for i, intercept in enumerate(mlp.intercepts_):
    print(f'Layer {i} intercepts:')
    print(intercept)
```
